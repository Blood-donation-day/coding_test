```python
# 카펫

def solution(brown, yellow):
    w, h = 3, 3
    
    while True:
        yel = (w-2) * (h-2)
        bro = w*h - yel
        
        if yellow == yel and bro == brown:
            return [w, h]
        elif yellow > yel and bro == brown:
            w -= 1
            h += 1
        else:
            w += 1
    
solution(24,24)
```

```python
# 영어 끝말잇기

def solution(n, words):
    words_set = set()
    words_set.add(words[0])
    
    for i in range(1, len(words)):
        if (words[i] in words_set) or (words[i-1][-1] != words[i][0]):
            return [(i%n)+1, (i//n) +1]
        
        words_set.add(words[i])
    
    return [0,0]
        
solution(3, ["tank", "kick", "know", "wheel", "land", "dream", "mother", "robot", "tank"])
```

```python
# 구명보트
from collections import deque as dq
def solution(people, limit):
    people.sort()
    people = dq(people)
    count = 0
    
    while len(people) > 0:
        people1 = people.pop()
        if len(people) > 0 and people[0] + people1 <= limit:
            people.popleft()
        count += 1

    return count

solution([70, 50, 80, 50, 80, 90], 100)
```

```python
# N개의 최소공배수

def solution(arr):
    arr.sort(reverse = True)
    result = arr[0]
    def gcd(a, b):
        if a % b == 0:
            return b
        else:
            return gcd(b, a % b)
    
    def lcm(a, b):
        return int(a * b / gcd(a, b))
    
    for i in range(1, len(arr)):
        result = lcm(result, arr[i])
    
    return result



solution([2,6,8,14])
```

```python
# 플레이 리스트 

def solution(genres, players):
    gen_dict = {}
    playlist = {}
    result = []
    for i in range(len(genres)):
        if genres[i] not in gen_dict:
            gen_dict[genres[i]] = players[i]
            playlist[genres[i]] = [[i,players[i]]]
        else:
            gen_dict[genres[i]] += players[i]
            playlist[genres[i]].append([i,players[i]])
    
    for gen in sorted(gen_dict.keys(), key=lambda x: gen_dict[x], reverse= True):
        playnum = 0
        for player in sorted(playlist[gen], key=lambda x:x[1], reverse=True):
            if playnum >= 2:
                break
            result.append(player[0])
            playnum += 1
            
    return result

solution(["classic", "pop", "classic", "classic", "pop", 'test'], [500, 600, 150, 800, 2500, 30000])
```

```python
# 기능 개발

def solution(progresses, speeds):
    result = []
    days = 0
    while len(progresses) > 0:
        
        days += int((100 - progresses[0]) // speeds[0])
        if (100 - progresses[0]) % speeds[0] != 0:
            days += 1
        count = 0
        for i in range(len(progresses)):
            progresses[i] += speeds[i] * days
            if progresses[i] < 100:
                break
        while len(progresses) > 0 and progresses[0] >= 100 :
            progresses.pop(0)
            speeds.pop(0)
            count += 1
        
        result.append(count)
    return result
    
solution([99, 98], [5, 1])
```

```python
# 프로세스 

from collections import deque
def solution(priorities, location):
    prior = deque(priorities)
    count = [i for i in range(len(priorities))]
    while 1:
        if prior[0] != max(prior):
            prior.append(prior.popleft())
            count.append(count.pop(0))

        else:
            prior.popleft()
            if count.pop(0) == location:
                return len(priorities) - len(count)
        
solution([2, 1, 3, 2], 2)
```

```python
# 다리를 지나는 트럭

from collections import deque
def solution(bridge_length, weight, truck_weights):
    bridge = deque([0] * bridge_length)
    time = 0
    sum_trucks = 0
    while truck_weights:
        sum_trucks -= bridge.popleft()
        if sum_trucks + truck_weights[0] <= weight:
            sum_trucks += truck_weights[0]
            bridge.append(truck_weights.pop(0))
        else: bridge.append(0)
        time += 1

    return time + len(bridge)

solution(2,10, [7,4,5,6])
```

```python
# 주식 가격 풀이
def solution(prices):
    stack = []
    answer = [0] * len(prices)
    
    for i in range(len(prices)):
        if stack != []:
            while stack != [] and stack[-1][1] > prices[i]:
                past, _ = stack.pop()
                answer[past] = i - past
        
        stack.append(i, prices[i])
    
    for i, s in stack:
        answer[i] = len(prices) -1 -i
    
    return answer
```

```python
# 더 맵게

import heapq as hp

def solution(scoville, K):
    hp.heapify(scoville) #리스트를 힙 자료형으로 만글기
    count = 0
    
    while scoville[0] < K:
        count += 1
        
        min_1 = hp.heappop(scoville)
        min_2 = hp.heappop(scoville)
        hp.heappush(scoville, min_1 + min_2 *2)

        if len(scoville) == 2 and (scoville[0] + scoville[1] * 2) < K:
            return -1 
    return count

solution([1,1,2,6], 27)
```

```python
# 숫자의 표현 

def solution(n):
    count = 1
    
    for i in range(1, n//2 +1):
        total = 0
        for j in range(i, n+2):
            if total > n:
                break
            elif total == n:
                count += 1
                break
            else:
                total += j

    return count

solution(15)
```

```python
# n^2 배열 자르기

def solution(n, left, right):
    # n ^ 2 로 배열을 NxN 만들기 
    arr = [[0 for i in range(n)] for j in range(n)]
    for i in range(n):
        for j in range(n):
            val = max(i, j) + 1
            arr[i][j] = val

    # 배열 잘라서 붙인다음 left right 슬라이스해서 리턴
    return sum(arr, [])[left:right+1]

solution(3,2,5)

# 이렇게 하니까 시간초과떠서 배열을 만들지 않고 계산하는 방법으로 변경

def solution(n, left, right):
    result = []
    
    for i in range(left, right+1):
        row, col = i // n, i % n
        result.append(max(row, col) + 1)
    return result

solution(3,2,5)
```

```python
# H-Index

def solution(citations):
    citations.sort(reverse=True)
    length = len(citations)
    for i in range(len(citations)):
        if citations[i] <= i+1:
            return max(citations[i], i)
    return length

solution([3, 0, 6, 1, 5,9,9,9,9,9,9])
```

```python
# 피로도 (완전탐색)

from itertools import permutations

def solution(k, dungeons):
    result = 0
    
    for pers in permutations(dungeons, len(dungeons)):
        count = 0
        stamina = k
        for n, s in pers:
            if n <= stamina:
                stamina -= s
                count += 1
        
        result = max(result, count)

    return result

solution(80, [[50,40],[50, 30],[30,10],[80,20],[30,10],[30,10],[30,10]])
```

```python
# 뉴스 클러스터링 다시 풀기

import re, math

def solution(str1, str2):
    str1 = [str1[i:i+2].lower() for i in range(0, len(str1)-1) if not re.findall('[^a-zA-Z]+', str1[i: i+2])]
    str2 = [str2[i:i+2].lower() for i in range(0, len(str2)-1) if not re.findall('[^a-zA-Z]+', str2[i:i+2])]
    
    gyo = set(str1) & set(str2)
    hap = set(str1) | set(str2)
    
    if len(hap) == 0:
        return 65536
    
    gyo_sum = sum([min(str1.count(gg), str2.count(gg)) for gg in gyo])
    hap_sum = sum([max(str1.count(hh), str2.count(hh)) for hh in hap])
    
    return math.floor((gyo_sum/hap_sum)*65536)

solution('aa+a154dw', 'AA55AWEA')
```
